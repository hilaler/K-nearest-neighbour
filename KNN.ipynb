```python
#import library
import pandas as pd
from math import sqrt

#reading excel file
data = pd.read_excel("https://github.com/hilaler/k-nearest-neighbour/blob/main/mobil.xls?raw=true")
data
```

```python
#check data if null then fill with 0
data = data.fillna(0)

#drop the 'Nama Mobil'
df = data.drop(['Nama Mobil'], axis = 1)
data
```

```python
def euclidean(a, b):
  return sqrt(sum([(a[i] - b[i])**2 for i in range(len(a))]))

def manhattan(a, b):
  return sum([abs(a[i] - b[i]) for i in range(len(a))])

def minkowski(a, b, p):
  return sum([abs(a[i] - b[i])**p for i in range(len(a))]) ** (1 / p)

def supremum(a, b):
  return max([abs(a[i] - b[i]) for i in range(len(a))])
```

```python
test = pd.DataFrame(pd.read_excel("https://github.com/hilaler/k-nearest-neighbour/blob/main/test.xlsx?raw=true"))
test = test.values
test
```

```python
def knn(x): 
  value = df.values

  euclidean_dis = []
  manhattan_dis = []
  minkowski_dis = []
  supremum_dis = []
  
  for i in value:
    outEuc = euclidean(i,test[0])
    outMan = manhattan(i,test[0])
    outMin = minkowski(i,test[0], 4)
    outSup = supremum(i,test[0])
    
    euclidean_dis.append(outEuc)
    manhattan_dis.append(outMan)
    minkowski_dis.append(outMin)
    supremum_dis.append(outSup)

  data["euclidean"] = euclidean_dis
  data["manhattan"] = manhattan_dis
  data["minkowski"] = minkowski_dis
  data["supremum"]  = supremum_dis
  
  euclidean_sort = data.sort_values(by=['euclidean'])[:x]
  manhattan_sort = data.sort_values(by=['manhattan'])[:x]
  minkowski_sort = data.sort_values(by=['minkowski'])[:x]
  supremum_sort = data.sort_values(by=['supremum'])[:x]
  
  print(euclidean_sort[['Nama Mobil', 'euclidean']])
  print()
  print(manhattan_sort[['Nama Mobil', 'manhattan']])
  print()
  print(minkowski_sort[['Nama Mobil', 'minkowski']])
  print()
  print(supremum_sort[['Nama Mobil', 'supremum']])
  return euclidean_sort, manhattan_sort, minkowski_sort, supremum_sort

euc, man, min, sup = knn(3)
euc["Nama Mobil"].to_excel('rekomendasi_euclidean.xls', index=False, header=False)
man["Nama Mobil"].to_excel('rekomendasi_manhattan.xls', index=False, header=False)
min["Nama Mobil"].to_excel('rekomendasi_minkowski.xls', index=False, header=False)
sup["Nama Mobil"].to_excel('rekomendasi_supremum.xls', index=False, header=False)
```
